```python
# Veritabanına sorgu gönderme fonksiyonu
def run_query(query):
    with driver.session() as session:
        result = session.run(query)
        return pd.DataFrame([r.values() for r in result], columns=result.keys())
```

```python
# Veritabanına sorgu gönderme fonksiyonu
def run_query(query):
    with driver.session() as session:
        session.run(query)
```

```python
# Fakülte düğümünü oluştur
def create_faculty_node():
    query = f"""
    MERGE (f:{fakulte} {{adi: '{fakulte_adi}'}})
    """
    run_query(query)
```

```python
# Bölüm düğümlerini oluştur ve fakülte düğümüyle ilişkilendir
def create_department_nodes():
    for bolum in bolumler:
        query = f"""
        MATCH (f:{fakulte} {{adi: '{fakulte_adi}'}})
        MERGE (b:Bolum {{adi: '{bolum}'}})
        MERGE (f)-[:BOLUM]->(b)
        """
        run_query(query)
```
