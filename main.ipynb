```python
def turkce_to_ingilizce(dizi):
    turkce_karakterler = "çÇğĞıİöÖşŞüÜ"
    ingilizce_karakterler = "cCgGiIoOsSuU"
    ceviri_tablosu = str.maketrans(turkce_karakterler, ingilizce_karakterler)
    
    sonuc_dizi = []
    for eleman in dizi:
        if any(char in turkce_karakterler for char in eleman):
            yeni_eleman = eleman.translate(ceviri_tablosu)
            sonuc_dizi.append(yeni_eleman)
        else:
            sonuc_dizi.append("yok")
    
    return sonuc_dizi

dizi = [
    "BASKIL MESLEK YUKSEKOKULU", "BINGOL HAYVAN SAGLIGI YUKSEKOKULU", "DEVLET KONSERVATUVARI",
    "DIS HEKIMLIGI FAKULTESI", "ECZACILIK FAKULTESI", "EGITIM BILIMLERI ENSTITUSU", "EGITIM FAKULTESI",
    "ELAZIG ORGANIZE SANAYI BOLGESI MESLEK YUKSEKOKULU", "FEN BILIMLERI ENSTITUSU", "FEN FAKULTESI",
    "IKTISADI VE IDARI BILIMLER FAKULTESI", "ILAHIYAT FAKULTESI", "ILETISIM FAKULTESI",
    "INSAN VE TOPLUM BILIMLERI FAKULTESI", "KARAKOCAN MESLEK YUKSEKOKULU", "KEBAN MESLEK YUKSEKOKULU",
    "KOVANCILAR MESLEK YUKSEKOKULU", "MIMARLIK FAKULTESI", "MUHENDISLIK FAKULTESI", "REKTORLUK",
    "SAGLIK BILIMLERI ENSTITUSU", "SAGLIK BILIMLERI FAKULTESI", "SAGLIK HIZMETLERI MESLEK YUKSEKOKULU",
    "SIVIL HAVACILIK YUKSEKOKULU", "SIVRICE MESLEK YUKSEKOKULU", "SOSYAL BILIMLER ENSTITUSU",
    "SOSYAL BILIMLER MESLEK YUKSEKOKULU", "SPOR BILIMLERI FAKULTESI", "SU URUNLERI FAKULTESI",
    "TEKNIK BILIMLER MESLEK YUKSEKOKULU", "TEKNOLOJI FAKULTESI", "TIP FAKULTESI",
    "UYGULAMALI BILIMLER FAKULTESI", "VETERINER FAKULTESI", "YABANCI DILLER YUKSEKOKULU"
]

sonuc = turkce_to_ingilizce(dizi)
for eleman in sonuc:
    print(eleman)
```
